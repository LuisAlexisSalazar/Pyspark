## **2. Concepto de Datalake**
---------------

### **2.1 Procesos ETL vs Procesos ELT**


- **Procesamiento ETL**: Proceso el cual consiste en la **Extracción** de los datos, su **Transformación** (limpieza, procesamineto) y su posterior **Carga** (base de datos, data mart, data warehourse, otros) con el fin de apoyar los análisis de negocio

  - Usados principalmente en arquitecturas de procesamiento de datos tradicionales

<br>

- **Procesos ELT**: Método que consiste en la **Extracción** de los datos, su **Carga** a los sistemas y su posterior **Trasnformación**

  - Usados para arquitecturas de procesamineto NoSQL tipo Big Data

### **3.1 Capa Landing**

- **Capa de aterrizaje de los datos en crudo**.
- Capa en donde se cargaran los diversos archivos para su posterior ingesta al data lake
- generalmente vienen formatos comunes como csv,json o xls

### **3.2 Capa Raw**

- Zona en donde los datos se encuentran sin un procesamiento mayor (dato en crudo)
- Datos no aptos para el consumo
- Usualmente cada campo es un string
- La capa raw contendrá datos que no serán necesariamente llevados a las posteriores capas del datalake.
- El beneficio de una Zona sin procesar es que permite ingerir nuevas fuentes rápidamente y establecer el linaje de los datos en Zonas más avanzadas en el procesamiento.

- generlamente pones datos del linaje para hacer la auditoria como fechas de inserción, fuentes de datos ... y generlamente un formato de Big Data como avro, parquet, orc.

### **3.3 Capa Silver o Master**

- Esta zona contiene estructuras que proporcionan un mayor valor que los datos sin procesar puros en la Zona sin procesar
- Datos disponibles en columnas con formato establecido y no simplemente como strings.
- En el caso de fuentes no estructuradas, es necesario realizar un procesamiento de los datos para ingresarlos en la Zona estructurada.
- Definir un formato donde ya no todo sea un string sino definir si es una estructura, número...

### **3.4 Capa Business o  Service Data**

- Capa donde se realizan lás lógicas necesarias a nuestra tabla para que pueda ser consumida por los usuarios finales

## **4. Pipeline de Datos**
--------------------------

- Construcción lógica que representa un proceso dividido en fases.
- Las pipelines de datos se caracterizan por definir el conjunto de pasos o fases y las tecnologías involucradas en un proceso de movimiento o procesamiento de datos.

- **Los pipeline nos ayudarán a realizar las transformaciones requeridas en las diversas zonas del datalake**

- Nos ayuda mover la información de un lado a otro de zona a zonas como de landing zone a row zone
